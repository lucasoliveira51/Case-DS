```python
tab1 = tabela.loc[(tabela['DT_VENCIMENTO']>='2019-03-01') & (tabela['DT_VENCIMENTO'] <= '2019-08-31')]
```

```python
ndf = tab1[tab1['ID_CONTA'].isin(tab['ID_CONTA'])]
ndf.loc[ndf['ID_CONTA']==205737]
```

```python
ndf[ndf['ID_CONTA']==755517]
```

```python
ndf1 = ndf.loc[ndf['DS_ROLAGEM']=='FX1']
```

```python
ndf2 = ndf1.groupby('ID_CONTA',as_index=False)['DS_ROLAGEM'].count()
```

```python
ndf2[ndf2['ID_CONTA']==1624908]
```

```python
df0 = tab1[tab1['ID_CONTA'].isin(tab['ID_CONTA'])]
df0.drop(['DT_VENCIMENTO','DS_ROLAGEM'],axis=1,inplace=True)
df_teste = pd.merge(df0,tab[['ID_CONTA','DT_VENCIMENTO','DS_ROLAGEM']],on='ID_CONTA')
df_teste = pd.merge(df_teste,ndf2,on='ID_CONTA')
```

```python
df0.head()
```

```python
a = df_teste.groupby('ID_CONTA',as_index=False).count()
a.drop(a.columns.difference(['ID_CONTA',"DT_VENCIMENTO"]),axis=1,inplace=True)
b = df_teste.groupby('ID_CONTA',as_index=False)['VL_FATURA'].mean()
df1 = pd.merge(a,b,on='ID_CONTA')
df1.rename(columns={'DT_VENCIMENTO': 'QTD_FATURAS_ULT_6M', 'VL_FATURA': 'VL_MEDIO_FATURA'}, inplace=True)
```

```python
df2 = pd.merge(df_teste,df1,on='ID_CONTA')
```

```python
df2.head()
```

```python
df2['DT_VENCIMENTO'] = pd.to_datetime(df2['DT_VENCIMENTO'])
df2.sort_values(by='DT_VENCIMENTO',ascending=True,inplace=True);df2.head()
```

```python
df2.drop_duplicates(subset=['ID_CONTA'],inplace=True,ignore_index=True)
```

```python
lista_cols = ['ID_CONTA','DS_ROLAGEM_x','DT_VENCIMENTO','QTD_FATURAS_ULT_6M','VL_MEDIO_FATURA','DS_ROLAGEM_y']

for cols in df2.columns:
  if cols in lista_cols:
    None
  else: 
    df2.drop([cols],axis=1,inplace=True) 

df2.rename(columns={'DS_ROLAGEM_x': 'DS_ROLAGEM', 'DT_VENCIMENTO': 'DT_VENCIMENTO',
                   'DS_ROLAGEM_y': 'QTD_FATURAS_ULT_6M_FX1'}, inplace=True)
df2 = df2.reindex(columns=['ID_CONTA','DS_ROLAGEM','DT_VENCIMENTO','QTD_FATURAS_ULT_6M','VL_MEDIO_FATURA','QTD_FATURAS_ULT_6M_FX1'])
```
