```python
base_2 = pd.read_csv("C://Users/Cliente/Desktop/estágio/Case Estagiário/Questão 2 - Base 2.txt", sep="	")
display(base_2)

sns.heatmap(base_2.corr(), annot=True, cmap="Wistia")
plt.show()
```

### Podemos observar que fazendo uma correlação entre os dados, obtemos que as correlações entre os meses são altas o que indica que, se o cliente tem mais parcelas está propenso a não pagar alguma fatura futuramente, uma solução para isso seria criar um plano da empresa para quando isso acontecer tomar uma medida como disponibilizar algum acordo com o cliente para efetuar o pagamento antecipado e receber algum tipo de cashback, fazendo com que o cliente pague no tempo certo, assim não correndo o risco de não pagar as faturas futuras.

```python
query = """select ID_CONTA, QTD_PARCELAMENTO_3M, QTD_PARCELAMENTO_6M, QTD_PARCELAMENTO_12M  from base_2 """
bs2=ps.sqldf(query, locals())
bs2.info()
```

```python
y = bs2["QTD_PARCELAMENTO_3M"]
x = bs2.drop("QTD_PARCELAMENTO_3M", axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)

modelo_regressaolinear = LinearRegression()
modelo_regressaolinear.fit(x_treino, y_treino)

previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
print(metrics.r2_score(y_teste, previsao_regressaolinear))
```

```python
y = bs2["QTD_PARCELAMENTO_6M"]
x = bs2.drop("QTD_PARCELAMENTO_6M", axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)

modelo_regressaolinear = LinearRegression()
modelo_regressaolinear.fit(x_treino, y_treino)

previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
print(metrics.r2_score(y_teste, previsao_regressaolinear))
```

```python
y = bs2["QTD_PARCELAMENTO_12M"]
x = bs2.drop("QTD_PARCELAMENTO_12M", axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)

modelo_regressaolinear = LinearRegression()
modelo_regressaolinear.fit(x_treino, y_treino)

previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
print(metrics.r2_score(y_teste, previsao_regressaolinear))
```

```python
tabela_auxiliar = pd.DataFrame()
tabela_auxiliar["y_teste"] = y_teste
tabela_auxiliar["Previsoes Regressao Linear"] = previsao_regressaolinear

plt.figure(figsize=(15,10))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```
